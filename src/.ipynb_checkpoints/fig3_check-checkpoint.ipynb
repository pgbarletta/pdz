```julia
using DataFrames
using Gadfly
using Compose

set_default_plot_size(30cm, 12cm)
```

```julia
home = "/home/german/labo/17/pdz/"
###
# 1
###
# Get ready
E_mat = Array{Float64, 2}(41, 400)
U_mat = Array{Float64, 2}(41, 400)
K_mat = Array{Float64, 2}(41, 400)
T_mat = Array{Float64, 2}(41, 400)

for i = 1:400
    # Read energies and temperature
    E_vec = readdlm(string(home, "run/lb/neq/data/1/outputs/", i, "etot.dat"))[:, 2]
    U_vec = readdlm(string(home, "run/lb/neq/data/1/outputs/", i, "eptot.dat"))[:, 2]
    K_vec = readdlm(string(home, "run/lb/neq/data/1/outputs/", i, "ektot.dat"))[:, 2]
    T_vec = readdlm(string(home, "run/lb/neq/data/1/outputs/", i, "temp.dat"))[:, 2]
    # Collect 'em
    E_mat[:, i] = E_vec
    U_mat[:, i] = U_vec 
    K_mat[:, i] = K_vec 
    T_mat[:, i] = T_vec
end

E_1 = reshape(mapslices(x -> mean(x), E_mat, 2), size(E_mat)[1])
U_1 = reshape(mapslices(x -> mean(x), U_mat, 2), size(E_mat)[1])
K_1 = reshape(mapslices(x -> mean(x), K_mat, 2), size(E_mat)[1])
T_1 = reshape(mapslices(x -> mean(x), T_mat, 2), size(E_mat)[1])

###
# 2
###
# Get ready
E_mat = Array{Float64, 2}(40, 400) 
U_mat = Array{Float64, 2}(40, 400)  
K_mat = Array{Float64, 2}(40, 400)
T_mat = Array{Float64, 2}(40, 400)

for i = 1:400
    # Read energies and temperature
    E_vec = readdlm(string(home, "run/lb/neq/data/2/outputs/", i, "etot.dat"))[:, 2]
    U_vec = readdlm(string(home, "run/lb/neq/data/2/outputs/", i, "eptot.dat"))[:, 2]
    K_vec = readdlm(string(home, "run/lb/neq/data/2/outputs/", i, "ektot.dat"))[:, 2]
    T_vec = readdlm(string(home, "run/lb/neq/data/2/outputs/", i, "temp.dat"))[:, 2]
    # Collect 'em
    E_mat[:, i] = E_vec
    U_mat[:, i] = U_vec 
    K_mat[:, i] = K_vec 
    T_mat[:, i] = T_vec
end


E_2 = reshape(mapslices(x -> mean(x), E_mat, 2), size(E_mat)[1])
U_2 = reshape(mapslices(x -> mean(x), U_mat, 2), size(E_mat)[1])
K_2 = reshape(mapslices(x -> mean(x), K_mat, 2), size(E_mat)[1])
T_2 = reshape(mapslices(x -> mean(x), T_mat, 2), size(E_mat)[1])

###
# 3
###
# Get ready
E_mat = Array{Float64, 2}(40, 400) 
U_mat = Array{Float64, 2}(40, 400)  
K_mat = Array{Float64, 2}(40, 400)
T_mat = Array{Float64, 2}(40, 400)

for i = 1:400
    # Read energies and temperature
    E_vec = readdlm(string(home, "run/lb/neq/data/3/outputs/", i, "etot.dat"))[:, 2]
    U_vec = readdlm(string(home, "run/lb/neq/data/3/outputs/", i, "eptot.dat"))[:, 2]
    K_vec = readdlm(string(home, "run/lb/neq/data/3/outputs/", i, "ektot.dat"))[:, 2]
    T_vec = readdlm(string(home, "run/lb/neq/data/3/outputs/", i, "temp.dat"))[:, 2]
    # Collect 'em
    E_mat[:, i] = E_vec
    U_mat[:, i] = U_vec 
    K_mat[:, i] = K_vec 
    T_mat[:, i] = T_vec
end

E_3 = reshape(mapslices(x -> mean(x), E_mat, 2), size(E_mat)[1])
U_3 = reshape(mapslices(x -> mean(x), U_mat, 2), size(E_mat)[1])
K_3 = reshape(mapslices(x -> mean(x), K_mat, 2), size(E_mat)[1])
T_3 = reshape(mapslices(x -> mean(x), T_mat, 2), size(E_mat)[1]);
```

```julia
# Compile in data frames
df_E = DataFrame(E = [ E_1 ; E_2 ; E_3 ], 
Time = [ collect(0:0.25:10) ; collect(10.25:0.25:20) ; collect(20.25:0.25:30) ],
Type = [ repeat([ "NVTs" ], inner = [41]) ; repeat([ "NVT"; "NVE" ], inner = [40]) ]);

df_U = DataFrame(U = [ U_1 ; U_2 ; U_3 ], 
Time = [ collect(0:0.25:10) ; collect(10.25:0.25:20) ; collect(20.25:0.25:30) ],
Type = [ repeat([ "NVTs" ], inner = [41]) ; repeat([ "NVT"; "NVE" ], inner = [40]) ]);


df_K = DataFrame(K = [ K_1 ; K_2 ; K_3 ], 
Time = [ collect(0:0.25:10) ; collect(10.25:0.25:20) ; collect(20.25:0.25:30) ],
Type = [ repeat([ "NVTs" ], inner = [41]) ; repeat([ "NVT"; "NVE" ], inner = [40]) ]);
```

```julia
plot(df_E, Geom.line, x="Time", y="E", color="Type",
Guide.title("E total"),
Theme(line_width = 2pt, grid_line_width = 1pt,
minor_label_font_size = 24pt, major_label_font_size = 30pt, key_label_font_size = 18pt,
key_title_font_size = 22pt))                                                       
```
