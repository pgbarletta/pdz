```python
from __future__ import print_function
%matplotlib inline
import mdtraj as md
import numpy as np
```

```python
# Get Ready
home = "/home/german/labo/17/pdz/"
traj = md.load(home+"run/lb/neq/data/tjump/Ttrajplb_2.nc", top = home+"top_files/dry_lb.prmtop")

cm_bandas_400 = []
aa1_bandas_400 = []
aa2_bandas_400 = []
aa3_bandas_400 = []
aa4_bandas_400 = []
aa5_bandas_400 = []
aa6_bandas_400 = []
aa7_bandas_400 = []
aa8_bandas_400 = []
```

```python
i = 0
for t in traj:
    t.xyz = t.xyz * 10
    # Get the first and last atoms of the ligand
    first = t.topology.residue(124).atom(0).index
    last = t.topology.residue(131).atom(t.topology.residue(131).n_atoms-1).index

    # Get the ligand topology
    df_lig_top = t.topology.to_dataframe()[0][first:last];
    df_lig_top.index = range(len(df_lig_top))
    df_lig_top['serial'] = df_lig_top['serial'] - 1944
    lig_top = md.Topology.from_dataframe(df_lig_top)

    # Get the ligand atoms coordinates
    lig_xyz = t.xyz[0, range(first, last)]

    # Construct a ligand trajectory
    lig_trj = md.Trajectory(lig_xyz, lig_top)

    # Get the ligand CM
    lig_cm_xyz = md.compute_center_of_mass(lig_trj)
    
    # Get the Calpha atoms indices and coordinates
    ca_indices = [a.index for a in t.topology.atoms if a.name == "CA" and a.index < first]
    ca_xyz = t.xyz[(0, tuple(ca_indices), )]

    # Get the Calpha atoms indices and coordinates
    lig_ca_indices = [a.index for a in t.topology.atoms if a.name == "CA" and a.index > first]
    lig_ca_xyz = t.xyz[(0, tuple(lig_ca_indices), )]
    
    ##################### CM #########################
    
    # Distance from ligand center of mass
    ca_xyz_cm = ca_xyz - lig_cm_xyz
    dist_cm = np.sqrt((ca_xyz_cm**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_cm)
    margin_hi = max(dist_cm) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi

    # Get the indices of the atoms that fall in each bin
    banda = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda = (dist_cm < bins[k]) & (dist_cm > bins[k-1])
        banda[k-1] = idx_banda.nonzero()

    # Store the atom indices for each band for this frame
    cm_bandas_400.append(banda)
    
    ##################### aa's #########################
    # aa1
    # Distance from ligand aa
    ca_xyz_aa1 = ca_xyz - lig_ca_xyz[0]
    dist_aa1 = np.sqrt((ca_xyz_aa1**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa1)
    margin_hi = max(dist_aa1) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa1 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa1 = (dist_aa1 < bins[k]) & (dist_aa1 > bins[k-1])
        banda_aa1[k-1] = idx_banda_aa1.nonzero()

    # Store the atom indices for each band for this frame
    aa1_bandas_400.append(banda_aa1)    

    # aa2
    # Distance from ligand aa
    ca_xyz_aa2 = ca_xyz - lig_ca_xyz[1]
    dist_aa2 = np.sqrt((ca_xyz_aa2**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa2)
    margin_hi = max(dist_aa2) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa2 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa2 = (dist_aa2 < bins[k]) & (dist_aa2 > bins[k-1])
        banda_aa2[k-1] = idx_banda_aa2.nonzero()

    # Store the atom indices for each band for this frame
    aa2_bandas_400.append(banda_aa2) 
    
    # aa3
    # Distance from ligand aa
    ca_xyz_aa3 = ca_xyz - lig_ca_xyz[2]
    dist_aa3 = np.sqrt((ca_xyz_aa3**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa3)
    margin_hi = max(dist_aa3) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa3 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa3 = (dist_aa3 < bins[k]) & (dist_aa3 > bins[k-1])
        banda_aa3[k-1] = idx_banda_aa3.nonzero()

    # Store the atom indices for each band for this frame
    aa3_bandas_400.append(banda_aa3) 

    # aa4
    # Distance from ligand aa
    ca_xyz_aa4 = ca_xyz - lig_ca_xyz[3]
    dist_aa4 = np.sqrt((ca_xyz_aa4**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa4)
    margin_hi = max(dist_aa4) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa4 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa4 = (dist_aa4 < bins[k]) & (dist_aa4 > bins[k-1])
        banda_aa4[k-1] = idx_banda_aa4.nonzero()

    # Store the atom indices for each band for this frame
    aa4_bandas_400.append(banda_aa4) 

    # aa5
    # Distance from ligand aa
    ca_xyz_aa5 = ca_xyz - lig_ca_xyz[4]
    dist_aa5 = np.sqrt((ca_xyz_aa5**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa5)
    margin_hi = max(dist_aa5) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa5 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa5 = (dist_aa5 < bins[k]) & (dist_aa5 > bins[k-1])
        banda_aa5[k-1] = idx_banda_aa5.nonzero()

    # Store the atom indices for each band for this frame
    aa5_bandas_400.append(banda_aa5) 

    # aa6
    # Distance from ligand aa
    ca_xyz_aa6 = ca_xyz - lig_ca_xyz[5]
    dist_aa6 = np.sqrt((ca_xyz_aa6**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa6)
    margin_hi = max(dist_aa6) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa6 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa6 = (dist_aa6 < bins[k]) & (dist_aa6 > bins[k-1])
        banda_aa6[k-1] = idx_banda_aa6.nonzero()

    # Store the atom indices for each band for this frame
    aa6_bandas_400.append(banda_aa6) 
    
    # aa7
    # Distance from ligand aa
    ca_xyz_aa7 = ca_xyz - lig_ca_xyz[6]
    dist_aa7 = np.sqrt((ca_xyz_aa7**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa7)
    margin_hi = max(dist_aa7) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa7 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa7 = (dist_aa7 < bins[k]) & (dist_aa7 > bins[k-1])
        banda_aa7[k-1] = idx_banda_aa7.nonzero()

    # Store the atom indices for each band for this frame
    aa7_bandas_400.append(banda_aa7)
    
    # aa8
    # Distance from ligand aa
    ca_xyz_aa8 = ca_xyz - lig_ca_xyz[7]
    dist_aa8 = np.sqrt((ca_xyz_aa8**2).sum(axis=1))

    # Create the array with the bins edges
    margin_lo = min(dist_aa8)
    margin_hi = max(dist_aa8) * 1.05
    step = (margin_hi - margin_lo) / 10
    bins = np.arange(0, margin_hi, step)
    bins[len(bins)-1] = margin_hi
    
    # Get the indices of the atoms that fall in each bin
    banda_aa8 = [None] * (len(bins)-1)
    for k in range(0, len(bins)-1):
        idx_banda_aa8 = (dist_aa8 < bins[k]) & (dist_aa8 > bins[k-1])
        banda_aa8[k-1] = idx_banda_aa8.nonzero()

    # Store the atom indices for each band for this frame
    aa8_bandas_400.append(banda_aa8) 
    

    ###### end
    i = i + 1
    print(i)
```
