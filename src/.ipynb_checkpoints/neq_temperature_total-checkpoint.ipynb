```julia
using DataFrames
using Gadfly
using Compose
set_default_plot_size(30cm, 12cm)
```

```julia
function meta_var(s::AbstractString,v::Any)
         s=Symbol(s) 
         @eval (($s) = ($v))
end
```

```julia
# Get ready
home = "/home/german/labo/17/pdz/"
corta_Taa_dir = "run/lb/neq/temperatura/total/corta/"
larga_Taa_dir = "run/lb/neq/temperatura/total/larga/"
const aa_cnt = 132
const corta_frame_cnt = 2000
const larga_frame_cnt = 180
```

```julia
###
# Corta
###
molde = string("corta_temp_total")
corta_T_total = Array{Float64, 1}(corta_frame_cnt)
T_mat = Array{Float64, 2}(corta_frame_cnt, 400)
for j = 1:400
    # Read temperature
    T_mat[:, j] = readdlm(string(home, corta_Taa_dir, j, molde))[:, 2]
end
# Get the averages
corta_T_total = reshape(mapslices(x -> mean(x), T_mat, 2), size(T_mat)[1]);

# Adapta la mtx y compila en dataframe
df_corta_T_total = DataFrame(T = corta_T_total, Time = collect(0.0005:0.0005:1));
```

```julia
###
# Larga
###
molde = string("larga_temp_total")
larga_T_total = Array{Float64, 1}(larga_frame_cnt)
T_mat = Array{Float64, 2}(larga_frame_cnt, 400)
for j = 1:400
    # Read temperature
    T_mat[:, j] = readdlm(string(home, larga_Taa_dir, j, molde))[:, 2]
end
# Get the averages
larga_T_total = reshape(mapslices(x -> mean(x), T_mat, 2), size(T_mat)[1]);

# Adapta la mtx y compila en dataframe
df_larga_T_total = DataFrame(T = larga_T_total, Time = collect(1.05:0.05:10));
```

# Corta

```julia
pdf_corta_T_total = plot(df_corta_T_total, Geom.line, x="Time", y="T",
Guide.title("NEQ run 0-1ps. PDZ+LIG  Temperature"),
Coord.cartesian(xmin = 0., xmax = 1, ymax = 500),
Theme(line_width = 2pt, grid_line_width = 1pt,
minor_label_font_size = 24pt, major_label_font_size = 30pt, key_label_font_size = 18pt,
key_title_font_size = 22pt))                                                      
```

## Larga

```julia
pdf_larga_T_total = plot(df_larga_T_total, Geom.line, x="Time", y="T",
Guide.title("NEQ run 1.05-10ps. PDZ+LIG  Temperature"),
Coord.cartesian(xmin = 1., xmax = 10, ymax = 500),
Theme(line_width = 2pt, grid_line_width = 1pt,
minor_label_font_size = 24pt, major_label_font_size = 30pt, key_label_font_size = 18pt,
key_title_font_size = 22pt))                                                      
```

## Todo

```julia
df_T_total = DataFrame(T = [ corta_T_total ; larga_T_total ],
    Time = [ collect(0.0005:0.0005:1) ; collect(1.05:0.05:10) ]);

pdf_T_total = plot(df_T_total, Geom.line, x="Time", y="T",
Guide.title("NEQ run 0-10ps. PDZ+LIG Temperature"),
Coord.cartesian(xmin = 0., xmax = 10, ymax = 500, ymin = 300),
Guide.yticks(ticks = collect(300:50:500)), Guide.xticks(ticks = collect(0:1:10)),
Theme(line_width = 2pt, grid_line_width = 1pt,
minor_label_font_size = 24pt, major_label_font_size = 30pt, key_label_font_size = 18pt,
key_title_font_size = 22pt))                                                      
```
